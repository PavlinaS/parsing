```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("user-data-dir=C:/Users/pvtar/AppData/Local/Google/Chrome/User Data/Profile 1")
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
parcel_id = pd.read_csv("C:/Users/pvtar/Downloads/ANTIOCH PINS.csv")
#parcel_id = parcel_id[60:]
parcel_id
```

```python
init = 'https://tax.lakecountyil.gov/search/commonsearch.aspx?mode=PARID'
```

```python
df = pd.DataFrame(columns=['PARID','Property Location','City/State/Zip','Neighborhood','Acres', 'Story Height',
                           'Architectural Style', 'Exterior Wall Cover', 'CDU', 'Bedrooms', 'Total AGLA', 'Attic', 'A/C',
                          'Asmt Year', 'Pay Year', 'Class', 'Land AV', 'Bldg AV', 'Total AV', 'Land MV', 'Bldg MV', 'Total MV',
                          'Taxable EAV', 'Reason Code', 'Sale Price', 'Sale Validity Code', 'Instrument #', 'Sales Type',
                          'Sale Date', 'Recorded Date', 'Instrument Type', 'Grantor/Seller', 'Grantee/Buyer'])
ids = []
location = []
city = []
neigh = []
acres = []

story = []
style = []
exter = []
cdu = []
beds = []
agla = []
attic = []
ac =[]

as_y = []
pa_y = []
clas = []
la_av = []
bl_av = []
to_av = []
la_mv = []
bl_mv = []
to_mv = []
tax = []
reas = []

s_pr = []
s_code = []
instr = []
s_type = []
s_date = []
rec_date = []
inst_type = []
grantor = []
grantee = []

DRIVER_PATH = 'C:/Users/pvtar/Documents/chromedriver.exe'
driver = webdriver.Chrome(executable_path=DRIVER_PATH, options=chrome_options)
driver.get(init)
#active the iframe and click the agree button
#WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.XPATH, "//iframe")))
agree = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="btAgree"]'))) 
agree.click()
#back to the main page
driver.switch_to.default_content()
inputElement = driver.find_element_by_id("inpParid")
inputElement.send_keys('0101300001')
inputElement.send_keys(Keys.ENTER)
for pin in parcel_id.PIN.head(50):
    print(pin)
    driver.get(init)
    inputElement = driver.find_element_by_id("inpParid")
    inputElement.send_keys(str(0) + str(pin))
    inputElement.send_keys(Keys.ENTER)
    PARIDs = driver.find_elements_by_class_name("DataletData")
    result = [{"text": parid.text} for parid in PARIDs]
    ids.append((str(0) + str(pin)))
    location.append(list(result[2].values())[0])
    city.append(list(result[4].values())[0])
    neigh.append(list(result[11].values())[0])
    acres.append(list(result[14].values())[0])
    time.sleep(3)
    driver.find_element_by_partial_link_text('Residential Building').click()
    ass_res = driver.find_elements_by_class_name("DataletData")
    result2 = [{"text": res.text} for res in ass_res]
    if not result2:
        print("Residential Building is empty")
        story.append('NaN')
        style.append('NaN')
        exter.append('NaN')
        cdu.append('NaN')
        beds.append('NaN')
        agla.append('NaN')
        attic.append('NaN')
        ac.append('NaN') 
    else:
        story.append(list(result2[3].values())[0])
        style.append(list(result2[4].values())[0])
        exter.append(list(result2[5].values())[0])
        cdu.append(list(result2[16].values())[0])
        beds.append(list(result2[17].values())[0])
        agla.append(list(result2[24].values())[0])
        attic.append(list(result2[25].values())[0])
        ac.append(list(result2[30].values())[0])    
    time.sleep(3)
    driver.find_element_by_partial_link_text('Values Current').click()
    val_res = driver.find_elements_by_class_name("DataletData")
    result3 = [{"text": res.text} for res in val_res]
    if not result3:
        print("Values Current is empty")
        as_y.append('NaN')
        pa_y.append('NaN')
        clas.append('NaN')
        la_av.append('NaN')
        bl_av.append('NaN')
        to_av.append('NaN')
        la_mv.append('NaN')
        bl_mv.append('NaN')
        to_mv.append('NaN')
        tax.append('NaN')
        reas.append('NaN')
    else:
        as_y.append(list(result3[0].values())[0])
        pa_y.append(list(result3[1].values())[0])
        clas.append(list(result3[2].values())[0])
        la_av.append(list(result3[3].values())[0])
        bl_av.append(list(result3[4].values())[0])
        to_av.append(list(result3[5].values())[0])
        la_mv.append(list(result3[6].values())[0])
        bl_mv.append(list(result3[7].values())[0])
        to_mv.append(list(result3[8].values())[0])
        tax.append(list(result3[9].values())[0])
        reas.append(list(result3[10].values())[0])
    time.sleep(3)
    driver.find_element_by_partial_link_text('Property Transfer History').click()
    trans_res = driver.find_elements_by_class_name("DataletData")
    result4 = [{"text": res.text} for res in trans_res]
    if not result4:
        print("Property Transfer History is empty")
        s_pr.append('NaN')
        s_code.append('NaN')
        instr.append('NaN')
        s_type.append('NaN')
        s_date.append('NaN')
        rec_date.append('NaN')
        inst_type.append('NaN')
        grantor.append('NaN')
        grantee.append('NaN')
    else:
        s_pr.append(list(result4[1].values())[0])
        s_code.append(list(result4[12].values())[0])
        instr.append(list(result4[3].values())[0])
        s_type.append(list(result4[11].values())[0])
        s_date.append(list(result4[9].values())[0])
        rec_date.append(list(result4[10].values())[0])
        inst_type.append(list(result4[6].values())[0])
        grantor.append(list(result4[7].values())[0])
        grantee.append(list(result4[8].values())[0])
    time.sleep(3)
#print(ids, location, city, neigh, acres)
```

```python
df['PARID'] = ids
df['Property Location'] = location
df['City/State/Zip'] = city
df['Neighborhood'] = neigh
df['Acres'] = acres

df['Story Height'] = story
df['Architectural Style'] = style
df['Exterior Wall Cover'] = exter
df['CDU'] = cdu
df['Bedrooms'] = beds
df['Total AGLA'] = agla
df['Attic'] = attic
df['A/C'] = ac

df['Asmt Year'] = as_y
df['Pay Year'] = pa_y
df['Class'] = clas
df['Land AV'] = la_av
df['Bldg AV'] = bl_av
df['Total AV'] = to_av
df['Land MV'] = la_mv
df['Bldg MV'] = bl_mv
df['Total MV'] = to_mv
df['Taxable EAV'] = tax
df['Reason Code'] = reas

df['Sale Price'] = s_pr
df['Sale Validity Code'] = s_code
df['Instrument #'] = instr
df['Sales Type'] = s_type
df['Sale Date'] = s_date
df['Recorded Date'] = rec_date
df['Instrument Type'] = inst_type
df['Grantor/Seller'] = grantor
df['Grantee/Buyer'] = grantee

df
```

```python
df.to_csv('first50.zip', index=False) 
```
